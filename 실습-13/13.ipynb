## outlier 

- 값이 다른 데이터 분포를 크게 벗어나는 데이터 
- 평균 등에 큰 영향을 줌
- 처리상의 오류일 수도 있고, 진짜 큰 값일 수도 있다  (몸무게 300kg vs. 500kg, 연매출 1조 vs. 8000조 ) 
- 통계적 정의 ( by John Tukey) 
    - 25% 값 - 1.5 X IQR 보다 작은 값 
    - 75% 값 + 1.5 X IQR 보다 큰 값
    

```python
# outlier 직접 구하기

Q1=df.대출금액.quantile(.25)
Q3=df.대출금액.quantile(.75)
u_border= Q3+(Q3-Q1)*1.5
l_border= Q1-(Q3-Q1)*1.5

print (l_border, u_border)
df[ df.대출금액 > u_border]
```

## 특이치,  결측치, 오류 데이터 

- 특이치 : 실제 데이터 값이 특이하게 크거나 작거나, 주로 numeric에서 
    - 원칙은 아웃라이어 기준으로 판단해야.. 
    - quantile (0.99) 값 등으로 대체 하기도 함
- 결측치 : 데이터 값이 없는 것 , NA라고도 함
    - 해당 레코드를 삭제하거나
    - 통계시 (통계란 column 단위로 이뤄짐) 해당값 빼고 계산
    - 대체 값 (평균, 중간값, 임의값, random 값) 을 만들어 넣기
    - column 별로 통계를 내는 것이기 때문에 결측치가 있어도 상관없다. 엉뚱한 값을 억지로 넣는 것 보다는.. 
            - 다만, paired value analysis (같은 레코드의 두 값을 비교 ) 하는 경우는 결측치가 있으면 안됨
- 오류 : 쓰레기 값이 들어있는 경우 
    - cleansing 이 필요: 원래 값을 살릴 수 있으면 살리고, 아니면 NA화 후 처리 
        - whitespace제거 등등
        - 같은 항목으로 count될 수 있게
    - 혈액형 A, B, O, AB 만 있어야 <= category의 경우 프로그램적으로 거를 수 있다

### 과제2 : 예금잔액, 적금 잔액에 대해  99.9% 상단코딩하시오

```python
# 예금잔액에 대한 99.9% 상단 값 계산
deposit_upper_limit = df['예금잔액'].quantile(0.999)

# 예금잔액 상위 값으로 조정
df.loc[df['예금잔액'] > deposit_upper_limit, '예금잔액'] = deposit_upper_limit

# 적금잔액에 대한 99.9% 상단 값 계산
savings_upper_limit = df['적금잔액'].quantile(0.999)

# 적금잔액 상위 값으로 조정
df.loc[df['적금잔액'] > savings_upper_limit, '적금잔액'] = savings_upper_limit

# 상자 그림(boxplot) 그리기
plt.boxplot([df['예금잔액'], df['적금잔액']], labels=['deposit', 'saving'], vert=False)
plt.show()
```

### 과제3 : 연체금액 전체에 대해 99%  상단코딩할 경우와 금액이 있는 경우에 대해서만 상단코딩할 경우를 비교하시오 

```python
# 99% 상단 값 계산 (전체 데이터 기준)
overdue_upper_limit_all = df['연체금액'].quantile(0.99)

# 99% 상단 값 계산 (금액이 있는 경우만 대상)
overdue_upper_limit_existing = df.loc[df['연체금액'] > 0, '연체금액'].quantile(0.99)

# 연체금액 상위 값으로 조정 (전체 데이터 기준)
df.loc[df['연체금액'] > overdue_upper_limit_all, '연체금액'] = overdue_upper_limit_all

# 연체금액 상위 값으로 조정 (금액이 있는 경우만 대상)
df.loc[(df['연체금액'] > 0) & (df['연체금액'] > overdue_upper_limit_existing), '연체금액'] = overdue_upper_limit_existing

# 출력 결과 비교
print("전체 데이터 기준 - 상단 코딩된 연체금액:")
print(df['연체금액'].describe())
print("\n금액이 있는 경우만 대상 - 상단 코딩된 연체금액:")
print(df.loc[df['연체금액'] > 0, '연체금액'].describe())
```

##  k 익명성

```python
import pandas as pd

# 파일 로딩

df=pd.read_csv ("data_org.csv", encoding="cp949", dtype={"식별번호":str , "성별": str, "우편번호":str})
df
```

## 준식별자를 구분하여 k 값 측정하며 맞추기

```python
# K 측정 
pd.set_option('display.max_rows',100)
pd.set_option('display.min_rows',None)

df.groupby(by=['성별', '나이','우편번호']).size().min() # K값
```

### k-익명성, 프라이버시 모델
- 안전도를 평가 하기 위한 수단이지 그 자체가 처리 방법은 아님
- K-익명성을 만족하기 위해
    - 범주화
    - 마스킹
    - 레코드 삭제
    - 상하단 코딩
    - 라운딩  등이 사용될 수 있음


```python
# K 만족을 위한 범주화, 마스킹, 레코드 삭제
# 나이 = 10세 단위

df['나이'] = df['나이'].round(-1)
df['나이']
df.groupby(by=['성별', '나이','우편번호']).size() # K값
```

```python
# K 및 EC 갯수, 크기 파악

def eval_K ( df ) :
    print ( "k값 : ", df.groupby(by=['성별', '나이', '우편번호']).size().min()    )
    print ("EC 갯수 : ", df.groupby(by=['성별', '나이', '우편번호']).size().count()  )
    #print ("EC 사이즈 별 개수\n", df.groupby(by=['성별', '나이', '우편번호']).size().value_counts() )    
    
eval_K(df)
```

```python
#  우편번호 mask 앞 3자리 

#df['우편번호']= df['우편번호'][ :3]+"**"
df['우편번호']=df['우편번호'].apply( lambda x : x[:2]+"***")

print (df['우편번호'])
```

### 과제3 : 다른 방식으로 k>5 달성하기

```python
# K값 측정
K_value = df.groupby(by=['성별', '나이', '우편번호']).size().min()
print("K값:", K_value)

# K값 만족을 위한 범주화
df['나이'] = df['나이'].round(-1)

# K 및 EC 갯수, 크기 파악
def eval_K(df):
    print("K값:", df.groupby(by=['성별', '나이', '우편번호']).size().min())
    print("EC 갯수:", df.groupby(by=['성별', '나이', '우편번호']).size().count())
    # print("EC 사이즈 별 개수\n", df.groupby(by=['성별', '나이', '우편번호']).size().value_counts())

eval_K(df)

# 우편번호 mask 앞 3자리
df['우편번호'] = df['우편번호'].apply(lambda x: x[:2] + "***")

print(df['우편번호'])

eval_K(df)

# 나이 하나의 분포를 봄
print(df["나이"].value_counts())

# 나이 상하단 코딩
df.loc[df.나이 > 70, '나이'] = 70
df.loc[df.나이 < 30, '나이'] = 30

print(df["나이"].value_counts())

eval_K(df)

# size 5 이상인 EC 보기
filtered_df = df.groupby(by=['성별', '나이', '우편번호']).filter(lambda x: len(x) >= 5)
print(filtered_df)

eval_K(filtered_df)

## K값: 5
## EC 갯수: 536
```
